```python
import numpy as np
from utils import *

np.random.seed(0)
# Define means and covariance matrices
means = [np.array([1, 1]), np.array([5, 10]), np.array([1, 11])]
covariances = [np.array([[5, 4], [4, 7]]) for _ in range(3)]
priors_X4 = [1/3, 1/3, 1/3]

# Generate datasets
X4, y4 = generate_data(means, covariances, priors_X4)

# Classify and compute errors
_, error_euclidean_X4 = classify_and_evaluate(X4, y4, euclidean_classifier, means)
_, error_mahalanobis_X4 = classify_and_evaluate(X4, y4, mahalanobis_classifier, means, covariances)
_, error_bayesian_X4 = classify_and_evaluate(X4, y4, bayesian_classifier, means, covariances, priors_X4)
# Print error rates
print(f"Euclidean Classifier Error (X4): {error_euclidean_X4:.4f}")
print(f"Mahalanobis Classifier Error (X4): {error_mahalanobis_X4:.4f}")
print(f"Bayesian Classifier Error (X4): {error_bayesian_X4:.4f}")

# Plot decision boundaries
plot_decision_boundaries(means, covariances, priors_X4, X4, y4, euclidean_classifier, "Euclidean Classifier on X4")
plot_decision_boundaries(means, covariances, priors_X4, X4, y4, mahalanobis_classifier, "Mahalanobis Classifier on X4")
plot_decision_boundaries(means, covariances, priors_X4, X4, y4, bayesian_classifier, "Bayesian Classifier on X4")
```
