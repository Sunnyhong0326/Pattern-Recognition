```python
import numpy as np
from utils import *

np.random.seed(0)
# Define means and covariance matrices
means = [np.array([1, 1]), np.array([4, 4]), np.array([1, 8])]
covariances = [np.eye(2) * 2 for _ in range(3)]
priors_X5 = [1/3, 1/3, 1/3]

# Generate datasets
X5, y5 = generate_data(means, covariances, priors_X5)

# Classify and compute errors
_, error_euclidean_X5 = classify_and_evaluate(X5, y5, euclidean_classifier, means)
_, error_mahalanobis_X5 = classify_and_evaluate(X5, y5, mahalanobis_classifier, means, covariances)
_, error_bayesian_X5 = classify_and_evaluate(X5, y5, bayesian_classifier, means, covariances, priors_X5)
# Print error rates
print(f"Euclidean Classifier Error (X5): {error_euclidean_X5:.4f}")
print(f"Mahalanobis Classifier Error (X5): {error_mahalanobis_X5:.4f}")
print(f"Bayesian Classifier Error (X5): {error_bayesian_X5:.4f}")

# Plot decision boundaries
plot_decision_boundaries(means, covariances, priors_X5, X5, y5, euclidean_classifier, "Euclidean Classifier on X5")
plot_decision_boundaries(means, covariances, priors_X5, X5, y5, mahalanobis_classifier, "Mahalanobis Classifier on X5")
plot_decision_boundaries(means, covariances, priors_X5, X5, y5, bayesian_classifier, "Bayesian Classifier on X6")
```
