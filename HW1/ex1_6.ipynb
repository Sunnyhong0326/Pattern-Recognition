```python
import numpy as np
from utils import *

np.random.seed(0)
# Define means and covariance matrices
means = [np.array([1, 1]), np.array([7, 7]), np.array([1, 15])]
covariances = [np.array([[1, 0], [0, 12]]), np.array([[2, 3], [3, 8]]), np.array([[2, 0], [0, 2]])]
priors_X6 = [1/3, 1/3, 1/3]

# Generate datasets
X6, y6 = generate_data(means, covariances, priors_X6)

# Classify and compute errors
_, error_euclidean_X6 = classify_and_evaluate(X6, y6, euclidean_classifier, means)
_, error_mahalanobis_X6 = classify_and_evaluate(X6, y6, mahalanobis_classifier, means, covariances)
_, error_bayesian_X6 = classify_and_evaluate(X6, y6, bayesian_classifier, means, covariances, priors_X6)
# Print error rates
print(f"Euclidean Classifier Error (X6): {error_euclidean_X6:.4f}")
print(f"Mahalanobis Classifier Error (X6): {error_mahalanobis_X6:.4f}")
print(f"Bayesian Classifier Error (X6): {error_bayesian_X6:.4f}")

# Plot decision boundaries
plot_decision_boundaries(means, covariances, priors_X6, X6, y6, euclidean_classifier, "Euclidean Classifier on X6")
plot_decision_boundaries(means, covariances, priors_X6, X6, y6, mahalanobis_classifier, "Mahalanobis Classifier on X6")
plot_decision_boundaries(means, covariances, priors_X6, X6, y6, bayesian_classifier, "Bayesian Classifier on X6")
```
