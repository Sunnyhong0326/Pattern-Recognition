```python
import numpy as np
from utils import *

np.random.seed(0)
# Define means and covariance matrices
means = [np.array([1, 1]), np.array([7, 14]), np.array([1, 16])]
covariances = [np.array([[4, 3],[3, 5]]) for _ in range(3)]
priors_X2 = [1/3, 1/3, 1/3]

# Generate datasets
X2, y2 = generate_data(means, covariances, priors_X2)

# Classify and compute errors
_, error_euclidean_X2 = classify_and_evaluate(X2, y2, euclidean_classifier, means)
_, error_mahalanobis_X2 = classify_and_evaluate(X2, y2, mahalanobis_classifier, means, covariances)
_, error_bayesian_X2 = classify_and_evaluate(X2, y2, bayesian_classifier, means, covariances, priors_X2)
# Print error rates
print(f"Euclidean Classifier Error (X2): {error_euclidean_X2:.4f}")
print(f"Mahalanobis Classifier Error (X2): {error_mahalanobis_X2:.4f}")
print(f"Bayesian Classifier Error (X2): {error_bayesian_X2:.4f}")

# Plot decision boundaries
plot_decision_boundaries(means, covariances, priors_X2, X2, y2, euclidean_classifier, "Euclidean Classifier on X2")
plot_decision_boundaries(means, covariances, priors_X2, X2, y2, mahalanobis_classifier, "Mahalanobis Classifier on X2")
plot_decision_boundaries(means, covariances, priors_X2, X2, y2, bayesian_classifier, "Bayesian Classifier on X2")
```
