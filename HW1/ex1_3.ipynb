```python
import numpy as np
from utils import *

np.random.seed(0)
# Define means and covariance matrices
means = [np.array([1, 1]), np.array([6, 8]), np.array([1, 13])]
covariances = [np.eye(2) * 6 for _ in range(3)]
priors_X3 = [1/3, 1/3, 1/3]

# Generate datasets
X3, y3 = generate_data(means, covariances, priors_X3)

# Classify and compute errors
_, error_euclidean_X3 = classify_and_evaluate(X3, y3, euclidean_classifier, means)
_, error_mahalanobis_X3 = classify_and_evaluate(X3, y3, mahalanobis_classifier, means, covariances)
_, error_bayesian_X3 = classify_and_evaluate(X3, y3, bayesian_classifier, means, covariances, priors_X3)
# Print error rates
print(f"Euclidean Classifier Error (X3): {error_euclidean_X3:.4f}")
print(f"Mahalanobis Classifier Error (X3): {error_mahalanobis_X3:.4f}")
print(f"Bayesian Classifier Error (X3): {error_bayesian_X3:.4f}")

# Plot decision boundaries
plot_decision_boundaries(means, covariances, priors_X3, X3, y3, euclidean_classifier, "Euclidean Classifier on X3")
plot_decision_boundaries(means, covariances, priors_X3, X3, y3, mahalanobis_classifier, "Mahalanobis Classifier on X3")
plot_decision_boundaries(means, covariances, priors_X3, X3, y3, bayesian_classifier, "Bayesian Classifier on X3")
```
