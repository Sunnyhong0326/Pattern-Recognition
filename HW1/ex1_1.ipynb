```python
import numpy as np
from utils import *

np.random.seed(0)
# Define means and covariance matrices
means = [np.array([1, 1]), np.array([8, 12]), np.array([1, 16])]
covariances = [np.eye(2) * 4 for _ in range(3)]
priors_X1 = [1/3, 1/3, 1/3]

# Generate datasets
X1, y1 = generate_data(means, covariances, priors_X1)

# Classify and compute errors
_, error_euclidean_X1 = classify_and_evaluate(X1, y1, euclidean_classifier, means)
_, error_mahalanobis_X1 = classify_and_evaluate(X1, y1, mahalanobis_classifier, means, covariances)
_, error_bayesian_X1 = classify_and_evaluate(X1, y1, bayesian_classifier, means, covariances, priors_X1)
# Print error rates
print(f"Euclidean Classifier Error (X1): {error_euclidean_X1:.4f}")
print(f"Mahalanobis Classifier Error (X1): {error_mahalanobis_X1:.4f}")
print(f"Bayesian Classifier Error (X1): {error_bayesian_X1:.4f}")

# Plot decision boundaries
plot_decision_boundaries(means, covariances, priors_X1, X1, y1, euclidean_classifier, "Euclidean Classifier on X1")
plot_decision_boundaries(means, covariances, priors_X1, X1, y1, mahalanobis_classifier, "Mahalanobis Classifier on X1")
plot_decision_boundaries(means, covariances, priors_X1, X1, y1, bayesian_classifier, "Bayesian Classifier on X1")
```
